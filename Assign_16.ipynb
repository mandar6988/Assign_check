# 17. Build a dataset with every possible (UserID, ProductID) pair (cross join)

```python
users['key'] = 1
products['key'] = 1
cross=pd.merge(users, products, on='key').drop('key', axis=1)
cross
```

```python
import itertools
user_ids = pd.unique(users['UserID'].values.ravel())
product_ids = pd.unique(products['ProductID'].values.ravel())
pairs = list(itertools.product(user_ids, product_ids))
df = pd.DataFrame(pairs, columns=['UserID', 'ProductID'])

df
```

# Determine how much quantity of each product was purchased by each user

```python
merged_df = transactions.merge(users, on='UserID', how='outer')
new_df =merged_df [["UserID", "ProductID", "Quantity"]]
sorted_df =new_df.sort_values(by="UserID")


print(sorted_df)
```

```python
uid = pd.DataFrame({"UserID":users["UserID"]})
tran = pd.DataFrame({"ProductID":products["ProductID"]})
uid['Key'] = 1 
tran['Key'] = 1 
data=pd.merge(uid,tran,how='outer',on="Key")
prod = pd.merge(data,transactions,how='left',on=['UserID','ProductID'])
quantity1 = prod.groupby(['UserID','ProductID'])['Quantity'].sum()
quantity1.reset_index().fillna(0)
quantity1
```
